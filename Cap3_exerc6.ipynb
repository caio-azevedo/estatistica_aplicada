## Com o objetivo de identificar o comportamento do consumidor inadimplente em relação aos seus hábitos de pagamento, foi realizada uma pesquisa com informações sobre a faixa etária do respondente e o grau de inadimplência. O objetivo é determinar se existe associação entre as variáveis. Com base no arquivo 'Inadimplência.csv', pede-se:

```python
# Importação dos pacotes necessários
import pandas as pd #manipulação de dados em formato de dataframe
from scipy.stats import chi2_contingency #estatística qui-quadrado
from scipy.stats.contingency import association #medidas de associação
import numpy as np #biblioteca para operações matemáticas multidimensionais
```

```python
url = "https://raw.githubusercontent.com/caio-azevedo/estatistica_aplicada/refs/heads/main/Semana%201/Inadimpl%C3%AAncia.csv"
df = pd.read_csv(url, sep=",")
# Visualização da base de dados 'Inadimplência'
df
```

a) Construa as tabelas de distribuição conjunta de frequências para as variáveis 'faixa_etaria' e 'inadimplencia' (frequências absolutas, frequências relativas em relação ao total geral, frequências relativas em relação ao total de cada linha, frequências relativas em relação ao total de cada coluna e frequências esperadas).

```python
tab_abs = pd.crosstab(df['faixa_etaria'], df['inadimplencia'], margins=True, margins_name='Total')
tab_abs
```

```python
tab_prop_all = pd.crosstab(df['faixa_etaria'], df['inadimplencia'], margins=True, margins_name='Total', normalize='all')
tab_prop_all
```

```python
tab_prop_row = pd.crosstab(df['faixa_etaria'], df['inadimplencia'], margins=True, margins_name='Total', normalize='index').round(3)
tab_prop_row
```

```python
tab_prop_col = pd.crosstab(df['faixa_etaria'], df['inadimplencia'], margins=True, margins_name='Total', normalize='columns').round(3)
tab_prop_col
```

```python
chi2, pvalor, df, expected = chi2_contingency(tab_abs)

# Tabela do Valores esperados
tab_esp = pd.DataFrame(expected, index = tab_abs.index, columns = tab_abs.columns)

tab_esp
```

```python
tab_cont = pd.crosstab(df['faixa_etaria'], df['inadimplencia'])
# Coeficiente de contingência
association(tab_cont, method='pearson')
```
